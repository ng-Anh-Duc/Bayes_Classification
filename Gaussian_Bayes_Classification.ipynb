```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/IRIS.csv')
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

species_encode = LabelEncoder()
df['species'] = species_encode.fit_transform(df['species'])
df.tail()
```

```python
import numpy as np

df = np.array(df, dtype=np.float64)
X = df[:,:4]
y = df[:,4]
X[:5,:]
```

```python
import math

def gaussian_function(data, mean, var):
    return (1.0/np.sqrt(2*math.pi*var)) * np.exp(-(data-mean)**2 / (2*var))
```

```python
gaussian_function(X[:,0], X[:,0].mean(), X[:,0].var())
```

```python
def compute_base_probability(data):
  species = np.unique(y)
  species_prop = []
  for s in species:
    species_prop.append(len(np.where(data==s)[0]) / len(data))
  return species_prop
print(np.unique(y))
compute_base_probability(y)
```

```python
def gaussian_bayes_classify(data, description=[5.2,4.6,3.7,2.8,1.3]):
  species = np.unique(y)
  species_prop = []
  for s in species:
    d = data[data[:,4] == s]
    prob = 1.0
    for i in range(len(description)):
      prob *= gaussian_function(description[i], d[:,i].mean(), d[:,i].var())
    species_prop.append(prob)
  max_prob = species_prop.index(max(species_prop))
  return species[max_prob]

gaussian_bayes_classify(df, description=[6.7,3,5.2,2.3])
```
